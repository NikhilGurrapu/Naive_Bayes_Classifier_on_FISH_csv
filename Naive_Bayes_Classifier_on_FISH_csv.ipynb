```python
xtrain,xtest,ytrain,ytest=train_test_split(df.iloc[:,1:].values,df['Species'].values,test_size=0.2)
```

## Scikit Learn Implementation

```python
from sklearn.naive_bayes import GaussianNB
```

```python
nb=GaussianNB()
nb.fit(xtrain,ytrain)
```

```python
print("Training R2 score: ",nb.score(xtrain,ytrain))
```

```python
ypred=nb.predict(xtest)
print("Accuracy_score: ",accuracy_score(ytest,ypred)*100)
```

```python
cm=confusion_matrix(ytest,ypred)
```

```python
print("Confusion Matrix:\n",cm)
print(classification_report(ytest,ypred))
print("Accuracy:",(cm[0][0]+cm[1][1])/(len(ytest))*100)
```
